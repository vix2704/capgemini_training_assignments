```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
# NORMALIZER
from sklearn.preprocessing import Normalizer
norm=Normalizer()
df = pd.DataFrame({
    'height': [162, 171, 155, 165],
    'weight': [63, 60, 50, 62],
    'speed': [5, 3, 7, 2]
})
data=norm.fit_transform(df)
df=pd.DataFrame(data,columns=df.columns)
df
```

```python
# STANDARD SCALER
from sklearn.preprocessing import StandardScaler
norm=StandardScaler()
df = pd.DataFrame({
    'height': [162, 171, 155, 165],
    'weight': [63, 60, 50, 62],
    'speed': [5, 3, 7, 2]
})
data=norm.fit_transform(df)
df=pd.DataFrame(data,columns=df.columns)
df
```
