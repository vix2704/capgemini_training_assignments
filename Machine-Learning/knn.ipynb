# KNN MODEL: CLASSIFICATION AND REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer, fetch_california_housing
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error,precision_score,recall_score, f1_score,mean_absolute_error
```

```python
cdata=load_breast_cancer()
cdata
```

```python
df=pd.DataFrame(cdata.data,columns=cdata.feature_names)
df['cancer']=cdata.target
df
```

```python
X=df.drop('cancer',axis=1)
y=df['cancer']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
model=KNeighborsClassifier(n_neighbors=10)
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
ypred
```

```python
cmatrix=confusion_matrix(ytest,ypred)
cmatrix
```

```python
print('Training efficiency', model.score(Xtrain,ytrain))
```

```python
print('Testing efficiency', model.score(Xtest,ytest))
```

```python
print('Accuracy', accuracy_score(ytest,ypred))
```

```python
print('Precision', precision_score(ytest,ypred))
```

```python
print('Recall', recall_score(ytest,ypred))
```

```python
print('F1 score', f1_score(ytest,ypred))
```

```python
# REGRESSOR
```

```python
hdata=fetch_california_housing()
hdata
```

```python
hf=pd.DataFrame(hdata.data,columns=hdata.feature_names)
hf['price']=hdata.target
hf
```

```python
X1=hf.drop('price',axis=1)
y1=hf['price']
```

```python
X1train,X1test,y1train,y1test=train_test_split(X1,y1,test_size=0.3, random_state=10)
```

```python
model1=KNeighborsRegressor(n_neighbors=2)
model1.fit(X1train,y1train)
```

```python
ypred1=model1.predict(X1test)
ypred1
```

```python
print('Training efficiency', model1.score(X1train,y1train))
```

```python
print('Testing efficiency', model1.score(X1test,y1test))
```

```python
print('MSE', mean_squared_error(y1test,ypred1))
```

```python
print('MAE',mean_absolute_error(y1test,ypred1))
```
