```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
df.head()
```

```python
pop=df['alcohol']
sample_size=500
sample=df['alcohol'].sample(sample_size)
sample_mean=sample.mean()
pop_mean=pop.mean()
error=pop_mean-sample_mean
```

```python
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop_mean,color='red',label='Population Mean')
plt.axhline(sample_mean,color='blue',label='Sample Mean')
plt.legend()
plt.show()
```

```python
#STRATIFIED SAMPLING

#calc mean for red wine n mean for white wine

red=df.groupby('type').get_group('red')
white=df.groupby('type').get_group('white')
# samples of red and white
samp_size=50
red_sample=red['alcohol'].sample(samp_size)
white_sample=white['alcohol'].sample(samp_size)
# mean
red_mean=red_sample.mean()
white_mean=white_sample.mean()
pop_mean=pop_mean-red_sample
print(pop_mean)
```

```python
# line chart
plt.plot(df['alcohol'])
plt.title('Alcohol Content Line Plot')
plt.xlabel('Sample Index')
plt.ylabel('Alcohol')
plt.show()

```

```python
avg_alcohol = df.groupby('quality')['alcohol'].mean()

plt.bar(avg_alcohol.index, avg_alcohol.values)
plt.title('Average Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Alcohol')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,10))   # make it bigger
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")  # limit decimals
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['alcohol'], df['quality'], alpha=0.6)
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Scatter Plot: Alcohol vs Quality')
plt.show()
```

```python
# box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['alcohol', 'density', 'quality', 'fixed acidity']])
plt.title('Box Plot of Selected Features')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,4))

# Histogram for Alcohol
plt.subplot(1,3,1)   # 1 row, 3 columns, 1st plot
plt.hist(df['alcohol'], bins=20, color='skyblue', edgecolor='black')
plt.title('Alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')

# Histogram for pH
plt.subplot(1,3,2)   # 2nd plot
plt.hist(df['pH'], bins=20, color='lightgreen', edgecolor='black')
plt.title('pH')
plt.xlabel('pH')

# Histogram for Density
plt.subplot(1,3,3)   # 3rd plot
plt.hist(df['density'], bins=20, color='salmon', edgecolor='black')
plt.title('Density')
plt.xlabel('Density')

plt.tight_layout()   # adjusts spacing so titles/labels don't overlap
plt.show()
```
