```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df=pd.DataFrame({'Gender': ['M','F','M','M','M']})
dc=pd.DataFrame({'colour': ['red','pink','blue']})

df['Gender']=le.fit_transform(df['Gender'])
dc['colour']=le.fit_transform(dc['colour'])
df
dc
```

```python
# onehot endoding for product
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder()
one=OneHotEncoder(sparse_output=False)
dp=pd.DataFrame({'Product':['laptop','tab','watch','tab','laptop','tab']})
data=one.fit_transform(dp[['Product']])
name=one.get_feature_names_out(['Product'])
display(data)
display(name)
```

```python
# ordinal encoding for review
from sklearn.preprocessing import OrdinalEncoder
ore=OrdinalEncoder(categories=[['Bad','Average','Good']])
df=pd.DataFrame({'review':['Bad','Average','Good','Bad','Good','Average']})
df['encodings']=ore.fit_transform(df[['review']])
display(df)
```
